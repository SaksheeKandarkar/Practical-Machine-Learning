```python
from pandas import read_csv
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout, Input
```

```python
df = pd.read_csv("C:\CDAC\ML\Topic13\car_price_prediction_regression.csv")
df.shape
```

```python
df_ohe = pd.get_dummies(df)
```

```python
df_ohe.shape
```

```python
df_ohe
```

```python
X=df_ohe.drop('Price',axis=1)
y=df_ohe['Price']
```

```python
import numpy as np
X=np.asarray(X).astype(np.float32)
```

```python
X.shape,y.shape
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=7)
```

# Practice NN

```python
n_features = X_train.shape[1]
```

```python
# define model
model1 = Sequential()
model1.add(Input((n_features,)))
model1.add(Dense(5, activation='relu'))# Batch Size:None
model1.add(Dense(1, activation='linear'))
```

```python
model1.summary()
```

```python
# compile the model
model1.compile(optimizer='adam', loss='mse',
               metrics='mae')#,'precision','recall','f1_score']
```

```python
# fit the model (traning)
history = model1.fit( X_train, y_train, epochs=100, batch_size=32,
                     validation_split=0.3, verbose=2,shuffle=True)
```
