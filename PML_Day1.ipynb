# Outlier Detection and Handling

```python
#Detect Outliers using Z-score
#For each column in X_ohe print number of rows where value is <-3 or >+3
```

```python
import numpy as np
# mean_X=np.mean(X_ohe,axis=1)
# std_X=np.std(X_ohe,axis=1)
# mean_X,std_X
# z_score=(X_ohe-mean_X)/std_X
# print(z_score)
for col in X_ohe.columns:
    outlier_count=X_ohe.loc[(X_ohe[col]<-3) | (X_ohe[col]>3),col].count()
    print(col,outlier_count)
```

```python
# For every column remove all outliers <-3 and >+3
X_ohe_clean=X_ohe.copy()
for col in X_ohe_clean.columns:
    print("Shape before dropping outliers",X_ohe_clean.shape)
    X_ohe_clean = X_ohe_clean.loc[ (X_ohe_clean[col]>=-3) & (X_ohe_clean[col]<=3) , : ]
    print("Shape after dropping outliers",X_ohe_clean.shape)
    
    
```

```python
# Impute the outliers
# For every column replace outliers <-3 and >+3 with valid values -3 or +3

X_ohe_clean = X_ohe.copy()
for col in X_ohe_clean.columns:
    X_ohe_clean.loc[ (X_ohe_clean[col] <-3 ),col]=-3
    X_ohe_clean.loc[ (X_ohe_clean[col] > 3 ),col] =3
```

# IQR Based Method for Outlier Handling

-Detect outliers using IQR method


```python
# IQR Based Method for outlier Handling
# For each column in X_ohe , print number of rows where value is < min_valid_val and > max_valid_val
X_ohe_clean = X_ohe.copy()
for col in X_ohe_clean.columns:
    q1,q3 = X_ohe_clean[col].quantile([0.25,0.75])
    IQR = q3 - q1
    min_valid_val = q1 - 1.5*IQR
    max_valid_val = q3 + 1.5*IQR
    outlier_count = X_ohe_clean.loc[(X_ohe_clean[col]<min_valid_val) | (X_ohe_clean[col]>max_valid_val), col].count()
    print(col,outlier_count)
```

```python
# Remove the outliers using IQR
X_ohe_clean=X_ohe.copy()
for col in X_ohe_clean.columns:
    q1,q3 = X_ohe_clean[col].quantile([0.25,0.75])
    IQR = q3 - q1
    min_valid_val = q1 - 1.5*IQR
    max_valid_val = q3 + 1.5*IQR
    print("Shape before dropping outliers",X_ohe_clean.shape)
    X_ohe_clean = X_ohe_clean.loc[ (X_ohe_clean[col]<=max_valid_val) & (X_ohe_clean[col]>=min_valid_val) , : ]
    print("Shape after dropping outliers",X_ohe_clean.shape)
    
```

```python
# Impute the outliers using IQR
X_ohe_clean = X_ohe.copy()
for col in X_ohe_clean.columns:
    q1,q3 = X_ohe_clean[col].quantile([0.25,0.75])
    IQR = q3 - q1
    min_valid_val = q1 - 1.5*IQR
    max_valid_val = q3 + 1.5*IQR
    
    X_ohe_clean.loc[ ( X_ohe_clean[col] > max_valid_val ) , col ] = max_valid_val
    X_ohe_clean.loc[ ( X_ohe_clean[col] < min_valid_val ) , col ] = min_valid_val

X_ohe_clean.max()>max_valid_val
X_ohe_clean.min()<min_valid_val
    
    
```
