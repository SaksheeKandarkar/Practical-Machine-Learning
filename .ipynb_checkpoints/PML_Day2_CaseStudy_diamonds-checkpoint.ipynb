# Train test Split

```python
from sklearn.model_selection import train_test_split
```

```python
diamonds_train,diamonds_test=train_test_split(diamonds_ohe,test_size=0.3,random_state=7)
```

# Agglomerative Hierrarchical Clustering

```python



import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10,7))
plt.title("Dendrograms")
dend = shc.dendrogram(shc.linkage(diamonds_train.iloc[:500,:],method='ward'))
plt.axhline(y=50,color='r',linestyle='--')

from sklearn.cluster import AgglomerativeClustering

ac = AgglomerativeClustering(2)

ac.fit(diamonds_train.iloc[:500,:])
```

```python
y_predict=ac.fit_predict(diamonds_train.iloc[:500,:])
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(diamonds_train.iloc[:500,:],y_predict)
```
