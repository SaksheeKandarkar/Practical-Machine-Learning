```python
X=diabetes1.drop('Diabetes_012',axis=1)
Y=diabetes1['Diabetes_012']
```

```python
# Standardize columns 
diabetes1_standard = (X - X.mean())/X.std()
```

# Outlier Handling

## Impute the outliers using z score based method

```python
X_clean = diabetes1_standard.copy()
for col in X_clean.columns:
    X_clean.loc[ (X_clean[col] <-3 ),col]=-3
    X_clean.loc[ (X_clean[col] > 3 ),col] =3
```

```python
X_min=X_clean.min() < -3
X_min
```

# Imbalanced data 

```python
#select all rows of class 1
X_c1 = X_clean.loc[Y==1,:]
Y_c1=Y[X_c1.index]
#select all rows of class 2
X_c2 = X_clean.loc[Y==2,:]
Y_c2=Y[X_c2.index]
#Undersample rows of class 0
X_c0=X_clean.loc[Y==0,:].sample(Y_c2.count())
Y_c0=Y[X_c0.index]
```

```python
X_un=pd.concat([X_c0,X_c1,X_c2])
Y_un=pd.concat([Y_c0,Y_c1,Y_c2])
X_un.shape,Y_un.shape
```

```python
Y_un.value_counts(normalize=True)
```

```python
from imblearn.combine import SMOTETomek
```

```python
smt=SMOTETomek(random_state=True)
```

```python
X_smt,Y_smt = smt.fit_resample(X_un,Y_un)
X_smt.shape,Y_smt.shape
```

```python
Y_smt.value_counts(normalize=True)
```

## Train Test Split
We have 3 variants of data

```python
from sklearn.model_selection import train_test_split
```

```python
X_smt_train,X_smt_test,Y_smt_train,Y_smt_test=train_test_split(X_smt,Y_smt,test_size=0.3,random_state=7,stratify=Y_smt)
X_smt_train.shape,X_smt_test.shape,Y_smt_train.shape,Y_smt_test.shape
```

# PCA Based Feature selection

```python
from sklearn.decomposition import PCA
from sklearn.tree import DecisionTreeClassifier
```

```python
pca = PCA(n_components=0.95,random_state=7)
pca.fit(X_smt_train,Y_smt_train)
X_train_smt_pca = pca.transform(X_smt_train)
X_test_smt_pca = pca.transform(X_smt_test)
```

```python
plt.plot(range(19),pca.explained_variance_ratio_)
```

# Decision Tree Based Feature selection

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=7)
dt.fit(X_smt_train,Y_smt_train)
```

```python
for imp,col in sorted(zip(dt.feature_importances_,X_smt_train.columns)):
  print(col,imp)
```

```python
mean_imp = sum(dt.feature_importances_) / len(dt.feature_importances_)
selected_col = X_smt_train.columns[ dt.feature_importances_ > mean_imp ]
X_train_smt_dt = X_smt_train.loc[:,selected_col]
X_test_smt_dt = X_smt_test.loc[:,selected_col]
X_train_smt_dt.shape , X_test_smt_dt.shape, Y_smt_train.shape,Y_smt_test.shape
```
