## Q3 b)

```python
from sklearn.model_selection import train_test_split
```

```python
# Split X and Y
X = df_ohe.drop(['Category'], axis=1)
Y = df_ohe['Category']
```

```python
# split into train and test datasets
X_train, X_test, y_train, y_test = train_test_split(X, Y,
                                                    test_size=0.3,
                                                    random_state=7,
                                                    stratify=Y)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout, Input
```

```python
from keras.optimizers import Adam
```

```python
n_features = X_train.shape[1]
n_features
```

```python
# define model
model1 = Sequential()
# Input layer
model1.add(Input((n_features,)))

#Hidden layers -3
model1.add(Dense(5, activation='relu'))# Batch Size:None
model1.add(Dense(4, activation='relu'))
model1.add(Dense(20, activation='relu'))

#Output layer
model1.add(Dense(1, activation='sigmoid'))
```

```python
model1.summary()
```

```python
# compile the model
model1.compile(optimizer='adam', loss='binary_crossentropy',
               metrics='precision')#,'precision','recall','f1_score']
```

```python
# fit the model (traning)
history = model1.fit( X_train, y_train, epochs=100, batch_size=430,
                     validation_split=0.3, verbose=0,shuffle=True)
```
