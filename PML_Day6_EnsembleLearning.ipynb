```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("C:\CDAC\ML\Topic 6\devnagari_db_csv\data.csv")
```

```python
df.columns
```

```python
df['character'].unique()
```

```python
df_data=df.loc[df['character'].isin(['character_21_pa',
       'character_22_pha','character_18_da',
       'character_19_dha','character_24_bha',
       'character_25_ma'])]
df_data.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df_data.iloc[:,0:-1]
Y=df_data.iloc[:,-1]
# Label encode y column
label_map={'character_21_pa':0,
       'character_22_pha':1,'character_18_da':2,
       'character_19_dha':3,'character_24_bha':4,
       'character_25_ma':5}
Y=Y.map(label_map)
X.shape,Y.shape
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=9)
```

# Multiclass classification problem

- Logistic Regression
- KNN
- Decision Tree
- Random Forest
- Adaboost
- XGBoost
- CatBoost
- LightGBM

```python

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from catboost import CatBoostClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
#pip install catboost
```

```python
rf=RandomForestClassifier(n_estimators = 500, max_depth=5,max_features='sqrt',oob_score=True,n_jobs=-1,random_state=7,max_samples=0.8)
rf.fit(X_train,Y_train)
```

# Adaboost

```python
ada = AdaBoostClassifier(estimator=DecisionTreeClassifier(random_state=7),n_estimators=100,learning_rate=1,random_state=7)
ada.fit(X_train,Y_train)
Y_pred=ada.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,Y_pred))
```
