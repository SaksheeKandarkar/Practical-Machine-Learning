# Feature Engineering

-create new features based on existing features

-In this data we will add two new columns / features age of the house and average price for a given zipcode

-age of the house= no. of years(year of sell (from date column) -yr built)

-avg price per zipcode :: pandas groupby

```python
df['date']=pd.to_datetime(df['date'],format='mixed')
```

```python
df['age'] = df['date'].dt.year - df['yr_built']
```

# Dealing with continuous variables with high skew
-Convert skewed variables into normal distribution

-We use Power transform

-Box-Cox transform # highly used in banks

-log transform

-square transform

-yeo johnson :: when there are negative values 

```python
from sklearn.preprocessing import PowerTransformer
```

```python
df['sqft_lot15_bc']=PowerTransformer(method='box-cox').fit_transform(df[['sqft_lot15']])
```

```python
df.skew()
```

```python
df.drop('sqft_lot15',axis=1,inplace=True)
```

```python
df.drop('zipcode',axis=1,inplace=True)
```

```python
df.shape
```

# X and Y split

```python
X= df.drop(['price'],axis=1)
Y=df['price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=7)
```

```python
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

# Linear Regression
# Regularization : LASSO ,Ridge

```python
from sklearn.linear_model import LinearRegression , Lasso, Ridge
```

```python
lr = LinearRegression()
lr.fit(X_train,Y_train)
```

```python
Y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
print(r2_score(Y_test,Y_pred),mean_squared_error(Y_test,Y_pred),mean_absolute_error(Y_test,Y_pred))
```

```python
ls = Lasso(alpha=1.0,random_state=7)
ls.fit(X_train,Y_train)
```

```python
Y_pred=ls.predict(X_test)
```

```python
print(r2_score(Y_test,Y_pred),mean_squared_error(Y_test,Y_pred),mean_absolute_error(Y_test,Y_pred))
```

```python
ls.coef_
```

```python
rd = Ridge(alpha=1,random_state=7)
rd.fit(X_train,Y_train)
```

```python
Y_pred=rd.predict(X_test)
```

```python
print(r2_score(Y_test,Y_pred),mean_squared_error(Y_test,Y_pred),mean_absolute_error(Y_test,Y_pred))
```

```python
rd.coef_
```

# Polynomial Features and Selection by Lasso ,Ridge

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
X_poly = PolynomialFeatures(degree=3).fit_transform(X)
X_poly
```

```python
X_poly.shape
```

```python
X_train_poly,X_test_poly,Y_train,Y_test = train_test_split(X_poly,Y,test_size=0.3,random_state=7)
```

```python
X_train_poly.shape,X_test_poly.shape,Y_train.shape,Y_test.shape
```

```python
lr=LinearRegression()
lr.fit(X_train_poly,Y_train)
```

```python
Y_pred=lr.predict(X_test_poly)
```

```python
print(r2_score(Y_test,Y_pred),mean_squared_error(Y_test,Y_pred),mean_absolute_error(Y_test,Y_pred))
```

```python
rd = Ridge(alpha=10000,random_state=7)
rd.fit(X_train_poly,Y_train)
Y_pred=rd.predict(X_test_poly)
```

```python
print(r2_score(Y_test,Y_pred),mean_squared_error(Y_test,Y_pred),mean_absolute_error(Y_test,Y_pred))
```

```python
ls = Lasso(alpha=10000,random_state=7)
ls.fit(X_train_poly,Y_train)
Y_pred=ls.predict(X_test_poly)
```

```python
print(r2_score(Y_test,Y_pred),mean_squared_error(Y_test,Y_pred),mean_absolute_error(Y_test,Y_pred))
```
