```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X = np.array([(4,1),(5,4),(6,3),(7,4),(2,-1),(-1,-2),(0,-3),(-1,-4)]).T
X[0,:],X[1,:]
X.mean(axis=1)
```

```python
plt.scatter(X[0,:],X[1,:])
plt.axhline(0,color='k')
plt.axvline(0,color='k')
plt.show
```

```python
#Centering the dataset
print(X,X.mean(axis=1).reshape(2,1))
X_centred=X-X.mean(axis=1).reshape(2,1)
print(X_centred)
plt.scatter(X_centred[0,:],X_centred[1,:])
plt.axhline(0,color='b')
plt.axvline(0,color='b')
```

```python
#Calculate the covariance matrix of X
cov_mat=X@X.T
cov_mat
```
