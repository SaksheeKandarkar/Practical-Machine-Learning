# Feature Selection 
-PCA

-KPCA

-Decision Tree Classifier

-RFE with Decision Tree Classifier

```python
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA,KernelPCA
```

# Split X and Y


```python
X=df_ohe.drop('Attrition_Flag',axis=1)
Y=df_ohe['Attrition_Flag']
X.shape,Y.shape
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=7,stratify=Y)
```

```python
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_selection import RFE
```

```python
scaler=StandardScaler()
scaler.fit(X_train)
```

```python
X_train.iloc[:,:]=scaler.transform(X_train.iloc[:,:])
X_test.iloc[:,:]=scaler.transform(X_test.iloc[:,:])
```

```python
X_train.mean(axis=0)
```

```python
pca= PCA(n_components=0.95,random_state=7)

pca.fit(X_train,Y_train)

X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
X_train_pca.shape,X_test_pca.shape
```

```python
kpca= KernelPCA(n_components=10,kernel='poly',degree=2,gamma=1,random_state=7)
kpca.fit(X_train)
X_train_kpca = kpca.transform(X_train)
X_test_kpca = kpca.transform(X_test)
```

```python
X_train_kpca.shape,X_test_kpca.shape
```

# Feature Selection Using DecisionTreeClassifier

```python
dt = DecisionTreeClassifier(criterion='entropy',random_state=7)
dt.fit(X_train,Y_train)
```

```python
rfe=RFE(DecisionTreeClassifier(random_state=7))
rfe.fit(X_train,Y_train)
```
